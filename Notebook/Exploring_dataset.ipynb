```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read dataset
input_df = pd.read_csv('./../Dataset/world_dataset.csv')
input_df.head()
```

```python
# Get length of dataset
print(f'Length of input dataset: {len(input_df)}\n')

# Get all columns name
columns_name = input_df.columns
columns_name
```

```python
input_df[input_df.isnull().any(axis=1)]
```

```python
# Check duplicated data and drop
input_df.duplicated().sum()
cleaned_duplicated = input_df.drop_duplicates()

# Convert to pandas NaN to handle error
cleaned_df = cleaned_duplicated.replace('nan', pd.NA)
print(f'Dataset length: {len(cleaned_df)}')
```

# **Comparative Analysis of GDP, Life Expectancy, and Unemployment Rate :**

```python
# Select relevant columns for comparative analysis
columns = ['Country', 'GDP', 'Life expectancy', 'Unemployment rate']
comparative_data = cleaned_df[columns]
#comparative_data.to_csv('../Report_materials/../Report_materials/comparative_data.csv', index=False)
comparative_data
```

```python
# Calculate descriptive statistics for each column
statistics = comparative_data.describe()
#statistics.to_csv('../Report_materials/statistics.csv', index=False)
statistics
```

```python
# Compare countries based on GDP
gdp_ranking = comparative_data.sort_values('GDP', ascending=False)
#gdp_ranking.to_csv('../Report_materials/gdp_ranking.csv', index=False)
gdp_ranking
```

```python
# Compare countries based on life expectancy
life_expectancy_ranking = comparative_data.sort_values('Life expectancy', ascending=False)
#life_expectancy_ranking.to_csv('../Report_materials/life_expectancy_ranking.csv', index=False)
life_expectancy_ranking
```

```python
top30_life_expectancy = life_expectancy_ranking.head(30)
plt.figure(figsize=(15, 6))  # Set the figure size
top30_life_expectancy.plot(x='Country', y='Life expectancy', kind='bar')
plt.title('Top 30 Countries - Life Expectancy')
plt.xlabel('Country')
plt.ylabel('Life Expectancy')
plt.xticks(rotation=90)
plt.tight_layout()  # Adjust spacing
plt.savefig('../Report_materials/top30_life_expectancy_graph.png')
```

```python
# Compare countries based on unemployment rate
unemployment_ranking = comparative_data.sort_values('Unemployment rate')
unemployment_ranking.to_csv('../Report_materials/unemployment_ranking.csv', index=False)
unemployment_ranking
```
